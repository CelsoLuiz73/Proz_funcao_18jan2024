```python
def media_aluno(nota1 = int, nota2 = int):
  media = (nota1 + nota2) / 2
  return media
```

```python
media_aluno(10, 7)
print("Média do aluno é: ", media_aluno)
```

```python
def ant_e_suc(num):
  ant = num - 1
  suc = num + 1
  return ant, suc

antecessor, sucessor = ant_e_suc(10)
print(antecessor)
print(sucessor)
```

```python
def indice_massa_corporal (peso = float, altura = float):
  imc = peso / (altura * altura)
  if (imc <= 18.5): return imc, "Magreza"
  elif (imc > 18.5) and (imc <= 24.9): return "Saudavel"
  elif (imc >= 25) and (imc <= 29.9): return "Sobrepeso"
  elif (imc > 30) and (imc <= 34.9): return "Obesidade grau 1"
  elif (imc > 35) and (imc <= 39.9): return "Obesidade severa grau 2"
  else : return "Seu indice é =", round(imc, 2), "=> Obesidade morbida grau 3"
  return imc
```

```python
indice = indice_massa_corporal(140.0, 1.73)
print(indice)
```
